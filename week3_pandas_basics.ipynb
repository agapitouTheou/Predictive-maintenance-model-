```python
# WEEK 3 PART 2: PANDAS BASICS
# Pandas = load, explore, analyze CSV data

import pandas as pd
import numpy as np

print("="*60)
print("PANDAS: WORKING WITH REAL EQUIPMENT DATA")
print("="*60)

# ===== 1. CREATE SAMPLE PREDICTIVE MAINTENANCE DATASET =====
print("\n1. CREATING SAMPLE EQUIPMENT DATASET")

# Real-world predictive maintenance data
data = {
    'engine_id': ['E001', 'E002', 'E003', 'E004', 'E005', 'E006', 'E007', 'E008'],
    'operating_hours': [1523, 2847, 450, 3200, 1100, 2950, 1800, 3500],
    'temperature_celsius': [85.2, 92.1, 78.5, 98.3, 81.2, 95.7, 88.4, 102.1],
    'pressure_psi': [2100, 2280, 1950, 2350, 2050, 2400, 2150, 2450],
    'vibration_g': [3.2, 4.1, 2.8, 5.2, 3.0, 4.8, 3.6, 5.5],
    'maintenance_required': ['No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes']
}

# Convert to DataFrame (table format)
df = pd.DataFrame(data)
print("\nDataFrame (first 5 rows):")
print(df.head())

# ===== 2. EXPLORE THE DATA =====
print("\n\n2. EXPLORING DATA")

print(f"\nDataset shape: {df.shape[0]} rows, {df.shape[1]} columns")
print(f"Columns: {list(df.columns)}")

print("\nData types:")
print(df.dtypes)

print("\nBasic statistics:")
print(df.describe())

# ===== 3. FILTER DATA (FIND PROBLEM EQUIPMENT) =====
print("\n\n3. FILTERING: FIND EQUIPMENT NEEDING MAINTENANCE")

# Equipment with high temperature (>95°C)
hot_equipment = df[df['temperature_celsius'] > 95]
print(f"\nEquipment with HIGH TEMPERATURE (>95°C):")
print(hot_equipment)

# Equipment with high vibration (>4.5G)
vibrating_equipment = df[df['vibration_g'] > 4.5]
print(f"\nEquipment with HIGH VIBRATION (>4.5G):")
print(vibrating_equipment)

# Equipment with BOTH problems (temp AND vibration)
problem_equipment = df[(df['temperature_celsius'] > 90) & (df['vibration_g'] > 4.0)]
print(f"\nEquipment with BOTH problems (temp>90 AND vibration>4.0):")
print(problem_equipment)

# ===== 4. COMPARE: MAINTENANCE vs NO MAINTENANCE =====
print("\n\n4. COMPARING EQUIPMENT WITH/WITHOUT MAINTENANCE ISSUES")

maintenance_yes = df[df['maintenance_required'] == 'Yes']
maintenance_no = df[df['maintenance_required'] == 'No']

print(f"\nEquipment NEEDING MAINTENANCE:")
print(f"  Average Temperature: {maintenance_yes['temperature_celsius'].mean():.1f}°C")
print(f"  Average Vibration:   {maintenance_yes['vibration_g'].mean():.2f}G")
print(f"  Average Hours:       {maintenance_yes['operating_hours'].mean():.0f}")

print(f"\nEquipment NOT NEEDING MAINTENANCE:")
print(f"  Average Temperature: {maintenance_no['temperature_celsius'].mean():.1f}°C")
print(f"  Average Vibration:   {maintenance_no['vibration_g'].mean():.2f}G")
print(f"  Average Hours:       {maintenance_no['operating_hours'].mean():.0f}")

# ===== 5. CALCULATE NEW FEATURES =====
print("\n\n5. CREATE NEW COLUMNS (ENGINEERED FEATURES)")

# Health score (YOUR MECHANICAL KNOWLEDGE!)
# High temp + high vibration + high pressure = UNHEALTHY
df['temp_normalized'] = (df['temperature_celsius'] - df['temperature_celsius'].min()) / (df['temperature_celsius'].max() - df['temperature_celsius'].min())
df['vibration_normalized'] = (df['vibration_g'] - df['vibration_g'].min()) / (df['vibration_g'].max() - df['vibration_g'].min())

# Simple health score (0-100, lower is better)
df['health_score'] = (df['temp_normalized'] + df['vibration_normalized']) / 2 * 100

print("\nEquipment with Health Scores (0-100, lower = healthier):")
print(df[['engine_id', 'temperature_celsius', 'vibration_g', 'health_score']])

# ===== 6. SORT AND RANK =====
print("\n\n6. RANK EQUIPMENT BY HEALTH (worst to best)")

df_sorted = df.sort_values('health_score', ascending=False)
print(df_sorted[['engine_id', 'health_score', 'maintenance_required']])

# ===== 7. BUSINESS INSIGHTS =====
print("\n\n7. PREDICTIVE MAINTENANCE INSIGHTS")

print(f"\nTotal equipment: {len(df)}")
print(f"Equipment needing maintenance: {len(maintenance_yes)} ({len(maintenance_yes)/len(df)*100:.0f}%)")
print(f"Equipment healthy: {len(maintenance_no)} ({len(maintenance_no)/len(df)*100:.0f}%)")

worst_equipment = df_sorted.iloc[0]
print(f"\nMOST CRITICAL EQUIPMENT: {worst_equipment['engine_id']}")
print(f"  Temperature: {worst_equipment['temperature_celsius']:.1f}°C (HIGH!)")
print(f"  Vibration: {worst_equipment['vibration_g']:.2f}G (HIGH!)")
print(f"  Health Score: {worst_equipment['health_score']:.1f}/100")
print(f"  Action: SCHEDULE MAINTENANCE IMMEDIATELY")

print("\n" + "="*60)
print("WHAT THIS MEANS FOR YOU:")
print("="*60)
print("""
This is REAL predictive maintenance work:
✓ Load equipment data from CSV
✓ Explore patterns and statistics
✓ Filter problem equipment automatically
✓ Calculate health scores
✓ Prioritize maintenance by risk

Your mechanical knowledge:
- You KNOW why high temp + vibration = failure risk
- You UNDERSTAND bearing wear symptoms
- You can VALIDATE that this analysis makes sense
- You can INTERPRET what the data is telling you

Next: Load REAL NASA turbofan dataset and do this for 100+ engines!
""")
```
